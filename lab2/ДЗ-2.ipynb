## Задача 2. Большие числа (1 балл)

Для натуральных чисел определена операция взятия факториала. Напомним, что это число вида $n! = 1 ⋅ 2 ⋅ \ldots ⋅ n$. При этом считают, что $0! = 1$.

Пользователь на вход вводит значение $n$. В ответе необходимо найти значение выражения
$$
1! + 2! + \dots + n!
$$

Пользоваться математической библиотекой math в этой задаче запрещено.

Пример ввода:
```
7
```

Пример вывода:
```
5913
```

````python
# Ваш код здесь
A = int(input())
S = 0
def fact(x): 
    m = 1
    while x > 0:
        m = m * x
        x = x - 1
    return m
while A > 0:
    S = S + fact(A)
    A = A - 1
if S == 0:
    S = 1
print(S)
````

## Задача 5. СКО (1 балл)

Пользователь вводит последовательность чисел, которая завершается нулем.

Определите для введенной последовательности стандартное нормальное отклонение (мера разброса данных), которая вычисляется по формуле

$$
\sigma = \sqrt{\frac{(x_1-s)^2+(x_2-s)^2+\ldots+(x_n-s)^2}{n-1}} $$

где
$$
s=\frac{x_1+x_2+\ldots+x_n}{n}
$$


Пример ввода:
```
10
1
1
1
1
1
0
```

Пример вывода:
```
3.67423461417
```

````python
# Ваш код здесь
x = int(input())
n = 0
l = []
while  x != 0:
    l.append(x)
    n += 1
    x = int(input())
s = sum(l)/n
sigma = 0
for i in l:
    sigma += (i-s)**2
sigma = (sigma/(n-1))**0.5
print(sigma)
````

По данному нечетному числу $n$ необходимо создать двумерный массив, образующий снежинку. После чего необходимо ее вывести в консоль. На главной и побочной диагоналях, а также в средней строке и среднем столбце стоят символы "*", в остальных ячейках - точки "."


Пример ввода:
```
9
```

Пример вывода:
```
* . . . * . . . *
. * . . * . . * .
. . * . * . * . .
. . . * * * . . .
* * * * * * * * *
. . . * * * . . .
. . * . * . * . .
. * . . * . . * .
* . . . * . . . *
```

````python
# Ваш код здесь
n = int(input())
for y in range(n):
    for x in range(n):
        c = "."
        if x == y:
            c = "*"
        if x == n//2:
            c = "*"
        if y == n//2:
            c = "*"
        if (n - 1) - x == y:
            c = "*"
        print(c, end=" ")
    print()
````

## Задача 8. Кегельбан (1 балл)

$N$ кеглей выставили в один ряд, занумеровав их слева направо числами от 1 до $N$. Затем по этому ряду бросили $K$ шаров, при этом $i$-й шар сбил все кегли с номерами от $l_i$ до $r_i$ включительно. Определите, какие кегли остались стоять на месте.

Программа получает на вход количество кеглей N и количество бросков $K$. Далее идет $K$ пар чисел $l_i$, $r_i$, при этом $1\le l_i\le r_i\le N$.

Программа должна вывести последовательность из $N$ символов, где $j$-й символ есть `I`, если $j$-я кегля осталась стоять, или `.`, если $j$-я кегля была сбита.

Пример ввода:
```
5 2
1 2
4 4
```

Пример вывода:
```
..I.I
```

Пояснение к примеру:
- до первого броска кегли были расположены так
```
IIIII
```
- после первого броска были сбиты кегли с первой по вторую
```
..III
```
- последний бросок выбил только четвертую кеглю
```
..I.I
```

````python
# Ваш код здесь
n, k = map(int, input().split())
a = ["I"] * n
for i in range(k):
    l, r = map(int, input().split())
    for j in range(l-1, r):
        a[j] = "."
print("".join(a))
````
